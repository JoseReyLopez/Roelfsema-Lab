```python
from glob import glob
import itertools
import numpy as np
```

```python
len(glob(f'Results_Bashivan_Vgg/**/*.pt', recursive = 1))

names = [i.split('/')[-1] for i in glob('Results_Bashivan_Vgg/**/*.pt', recursive = 1)]

rois = []
vgg_layers = []

for name in names:
    split = name.split('_')
    roi, vgg_layer = split[2:4]
    rois.append(int(roi))
    vgg_layers.append(int(vgg_layer))
    #print(name)
    #print(split[2:4], split[split.index('loss')+1])

rois = np.unique(rois)
vgg_layers = np.unique(vgg_layers)

combinations = []
for roi, vgg_layer in itertools.product(rois, vgg_layers):
    #print(roi, vgg_layer)
    files = glob(f'Results_Bashivan_Vgg/**/*_{roi}_{vgg_layer}_*.pt', recursive = 1)
    loss = []
    for file in files:
        file_name = file.split('/')[-1].split('_')
        loss_value = file_name[file_name.index('loss')+1]
        if len(loss_value)!=0:
            loss.append(loss_value)
    if len(loss)!=0:    
        loss = list(map(float, loss))
        #print(roi, vgg_layer, np.min(loss))
        combinations.append([roi, vgg_layer, np.min(loss)])
    #print('\n\n\n')






rois = np.unique(np.array(combinations)[:, 0])
vgg_layers = np.unique(np.array(combinations)[:, 1])

result_bashivan_vgg = np.zeros((len(rois), len(vgg_layers)))
result_bashivan_vgg_copy = np.copy(result_bashivan_vgg)

combinations = np.array(combinations)



for roi, vgg_layer in itertools.product(rois, vgg_layers):
    position = np.intersect1d(np.where(combinations[:, 0] == roi), np.where(combinations[:, 1] == vgg_layer))
    
    if len(position) != 0:
        result_bashivan_vgg[np.where(rois == roi)[0], np.where(vgg_layers == vgg_layer)[0]] = combinations[position[0], 2]


result_bashivan_vgg_copy = np.copy(result_bashivan_vgg)
result_bashivan_vgg_copy[np.where(result_bashivan_vgg==0)] = 100

np.argmin(result_bashivan_vgg_copy, 1)
```

```python
# What a horrible piece of code with so many tries and excepts, 
# but if it works, it works (for now...)


import torch
from autoload_models import *



def layer_minimization_per_roi(model, CNN):
    
    '''
    A model ['Cadena', 'Bashivan'] and a pretrained cnn ['Vgg', 'Inception'] are provided
    and the function returns which is the layer (for Vgg [0,2,5,7,10,12,14...], for Inception
    ['conv2d0', 'conv2d1'...])
    The associated roi is encoded in the possition of the list. If no model was trained for 
    a given roi, that position will contain a None.
    '''


    assert model in ['Cadena', 'Bashivan']
    assert CNN in ['Vgg', 'Inception']

    ########


    neurons_per_roi = {1: 43, 2: 62, 3: 55, 4: 50, 5: 51, 7: 53, 8: 60, 9: 43, 10: 44, 11: 53, 12: 29, 13: 17, 14: 24, 15: 59, 16: 24}

    vgg_layers = [0,2,5,7,10,12,14,16,19,21,23,25,28,30,32,34]
    inception_layers  = ['conv2d0', 'conv2d1', 'conv2d2', 'mixed4a', 'mixed4b', 'mixed4c', 'mixed4d', 'mixed4e', 'mixed5a', 'mixed5b']
    rois = list(range(1,17))


    ########



    if CNN == 'Vgg':
        layers = vgg_layers
    if CNN == 'Inception':
        layers = inception_layers

    max_layer_per_roi = []

    for roi in rois:

        active_layers = []
        losses_per_layer = []
        for layer in layers:
            #print(roi, layer)
            model_name = None
            if model == 'Cadena' and CNN == 'Vgg':
                try:
                    model_name = FindLowestCadenaVgg(roi, layer)
                except:
                    pass
            if model == 'Cadena' and CNN == 'Inception':
                try:
                    model_name = FindLowestCadenaInception(roi, layer)
                except:
                    pass
            if model == 'Bashivan' and CNN == 'Vgg':
                try:
                    model_name = FindLowestBashivanVgg(roi, layer)
                except:
                    pass
            if model == 'Bashivan' and CNN == 'Inception':
                try:
                    model_name = FindLowestBashivanInception(roi, layer)
                except:
                    pass
            
            if model_name != None:
                #print(roi, layer)
                splitted_name = model_name.split('/')[-1].rsplit('.', 1)[0].split('_')
                losses_per_layer.append(float(splitted_name[splitted_name.index('loss')+1]))
                active_layers.append(layer)

        #print(roi, losses_per_layer, active_layers)
        if len(losses_per_layer)>0:
            max_layer_per_roi.append(active_layers[np.argmin(losses_per_layer)])
        else:
            max_layer_per_roi.append(None)
            
    return max_layer_per_roi
```
