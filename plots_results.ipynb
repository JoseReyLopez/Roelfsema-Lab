# Results Bashivan VGG

```python
len(glob('Results_Bashivan_Vgg/**/*.pt', recursive = 1))
```

```python
names = [i.split('/')[-1] for i in glob('Results_Bashivan_Vgg/**/*.pt', recursive = 1)]
```

```python
rois = []
vgg_layers = []
```

```python
for name in names:
    split = name.split('_')
    roi, vgg_layer = split[2:4]
    rois.append(int(roi))
    vgg_layers.append(int(vgg_layer))
    print(name)
    print(split[2:4], split[split.index('loss')+1])

rois = np.unique(rois)
vgg_layers = np.unique(vgg_layers)
```

```python
combinations = []
for roi, vgg_layer in itertools.product(rois, vgg_layers):
    print(roi, vgg_layer)
    files = glob(f'Results_Bashivan_Vgg/**/*_{roi}_{vgg_layer}_*.pt', recursive = 1)
    loss = []
    for file in files:
        file_name = file.split('/')[-1].split('_')
        loss_value = file_name[file_name.index('loss')+1]
        if len(loss_value)!=0:
            loss.append(loss_value)
    if len(loss)!=0:    
        loss = list(map(float, loss))
        print(roi, vgg_layer, np.min(loss))
        combinations.append([roi, vgg_layer, np.min(loss)])
    print('\n\n\n')
```

```python
combinations
```

```python
rois = np.unique(np.array(combinations)[:, 0])
vgg_layers = np.unique(np.array(combinations)[:, 1])

result_bashivan_vgg = np.zeros((len(rois), len(vgg_layers)))
result_bashivan_vgg_copy = np.copy(result_bashivan_vgg)

combinations = np.array(combinations)
```

```python
for roi, vgg_layer in itertools.product(rois, vgg_layers):
    position = np.intersect1d(np.where(combinations[:, 0] == roi), np.where(combinations[:, 1] == vgg_layer))
    
    if len(position) != 0:
        result_bashivan_vgg[np.where(rois == roi)[0], np.where(vgg_layers == vgg_layer)[0]] = combinations[position[0], 2]
```

```python
result_bashivan_vgg_copy = np.copy(result_bashivan_vgg)
result_bashivan_vgg_copy[np.where(result_bashivan_vgg==0)] = 100
```

```python
np.argmin(result_bashivan_vgg_copy, 1)
```

```python
plt.figure(figsize = (20,20))

plt.imshow(result_bashivan_vgg, cmap = mpl.cm.RdYlGn_r)
plt.xticks(range(16), map(str, map(int, vgg_layers)))
plt.yticks(range(14), map(str, map(int, rois)))
plt.xlabel('Vgg layers', fontsize = 20)
plt.ylabel('ROIS', fontsize = 20)
plt.clim(0.01, 0.2)
cbar = plt.colorbar(fraction = 0.04)

plt.suptitle('Fitness of Bashivan and VGG19', fontsize = 35)


lw = 1
for x_value in range(0, 15):
    plt.plot((x_value+.5,x_value+.5), (-.5,14-.5), linewidth = lw, color = 'k')
for y_value in range(0, 15):
    plt.plot((0-.5, 16-.5), (y_value-.5,y_value-.5), linewidth = lw, color = 'k')


def rectangle(x, y, color = 'r', lw = 5):
    plt.plot((x-.5,x+.5), (y-.5, y-.5), linewidth = lw, color = color)
    plt.plot((x-.5,x+.5), (y+.5, y+.5), linewidth = lw, color = color)

    plt.plot((x-.5,x-.5), (y+.5, y-.5), linewidth = lw, color = color)
    plt.plot((x+.5,x+.5), (y+.5, y-.5), linewidth = lw, color = color)

result_bashivan_vgg_copy = np.copy(result_bashivan_vgg)
result_bashivan_vgg_copy[np.where(result_bashivan_vgg==0)] = 100

for roi, best_layer in enumerate(np.argmin(result_bashivan_vgg_copy, 1)):
    rectangle(best_layer, roi, 'blue')

for x_i, y_i in itertools.product(range(len(rois)), range(len(vgg_layers))):
    if np.round(result_bashivan_vgg[x_i, y_i], 4) != 0:
        plt.text(y_i-.3, x_i, np.round(result_bashivan_vgg[x_i, y_i], 4))

plt.savefig('/home/jose/Desktop/Bashivan_VGG.png', dpi = 150)
```

# Results Bashivan Inception

```python
len(glob('Results_Bashivan_Inception/**/*.pt', recursive = 1))
```

```python
names = glob('Results_Bashivan_Inception/**/*.pt', recursive = 1)
names = [i.split('/')[-1] for i in names]
```

```python
rois    = []
layers = []

for name in names:
    roi, layer = name.split('_')[2:4]
    rois.append(int(roi))
    layers.append(layer)

rois = np.unique(rois)
layers = np.unique(layers)

print(rois, '\n', layers)
```

```python
combinations = []

for roi, layer in itertools.product(rois, layers):
    model_files = glob(f'Results_Bashivan_Inception/**/*_{roi}_{layer}_*.pt', recursive=1)
    loss = []

    for model_file in model_files:
        file_name = model_file.split('/')[-1].split('_')
        loss.append(float(file_name[file_name.index('loss')+1]))

    combinations.append([roi, layer, np.min(loss)])
```

```python
combinations = np.array(combinations)
```

```python
combinations.shape
```

```python
results_bashivan_inception = np.zeros((len(rois), len(layers)))
results_bashivan_inception_copy = np.copy(results_bashivan_inception)
```

```python
for roi, layer in itertools.product(rois, layers):
    print(roi, layer)
    #print(np.where(combinations[:, 0] == str(roi)), np.where(combinations[:, 1] == layer))
    position = np.intersect1d(np.where(combinations[:, 0] == str(roi)), np.where(combinations[:, 1] == layer))

    if len(position) != 0:
        #print(np.where(rois == roi)[0][0], '    ', np.where(layers == layer)[0][0], '    ', combinations[position[0], 2])
        results_bashivan_inception[np.where(rois == roi)[0][0], np.where(layers == layer)[0][0]] = combinations[position[0], 2]
    
```

```python
plt.figure(figsize = (20,20))

plt.imshow(results_bashivan_inception, cmap = mpl.cm.RdYlGn_r)
plt.yticks(range(len(rois)), map(str, rois))
plt.xticks(range(len(layers)), layers)
plt.xlabel('Inception Layers', fontsize = 20)
plt.ylabel('ROIS', fontsize = 20)

plt.suptitle('Fitness of Bashivan and InceptionV1', fontsize = 35)


def rectangle(x, y, color = 'r', lw = 5):
    plt.plot((x-.5,x+.5), (y-.5, y-.5), linewidth = lw, color = color)
    plt.plot((x-.5,x+.5), (y+.5, y+.5), linewidth = lw, color = color)

    plt.plot((x-.5,x-.5), (y+.5, y-.5), linewidth = lw, color = color)
    plt.plot((x+.5,x+.5), (y+.5, y-.5), linewidth = lw, color = color)

best_layers = np.argmin(results_bashivan_inception, 1)

for n, best_layer in enumerate(best_layers):
    rectangle(best_layer,n , 'blue')

lw = 1
for x_value in range(0, 10):
    plt.plot((x_value+.5,x_value+.5), (-.5,14-.5), linewidth = lw, color = 'k')
for y_value in range(0, 14):
    plt.plot((0-.5, 10-.5), (y_value-.5,y_value-.5), linewidth = lw, color = 'k')

for x_i, y_i in itertools.product(range(len(rois)), range(len(layers))):
    plt.text(y_i-.2, x_i, np.round(results_bashivan_inception[x_i, y_i], 4))

plt.colorbar()

plt.savefig('/home/jose/Desktop/Bashivan_Inception.png', dpi = 150)
```
